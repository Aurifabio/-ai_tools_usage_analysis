# Analysis of AI Tools Usage Among High School Students
##### Source: Kaggle Dataset

In this project, we will perform an exploratory data analysis (EDA) using the dataset "AI Tools Usage Among Global High School Students". With the growing adoption of artificial intelligence (AI) tools, students worldwide are starting to integrate these technologies into their educational routines.

Our goal is to understand how high school students across different countries are using AI tools, exploring patterns related to country, gender, grade level, usage frequency, and perceived impact of these technologies.

- Key questions we aim to answer:
What is the students' level of familiarity with AI?

- What are the most common purposes for using AI?

- Are there usage differences across countries or genders?

- How do students perceive the impact of AI (positive or negative)?

```python
# importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('global_ai_tools_students_use.csv')
df.head()
```

```python
# The function displays descriptive statistics from the DataFrame, including means, quartiles, 
# and standard deviation for numbers, and counts and mode for categories.
df.describe(include='all')
```

```python
# Shows the size of the dataset (500 rows and 22 columns)
df.shape
```

```python
# sum the number of null numbers (NaN) per column
df.isnull().sum()
```

```python
# Replaces missing values (NaN) in all columns that start with 'usefulness' with 0.
# This ensures that non-responses about the usefulness of AI tools are treated as zero,
# indicating no perceived usefulness or no usage.
fill_nulls = [c for c in df.columns if c.startswith('usefulness')]
df[fill_nulls] = df[fill_nulls].fillna(0)
```

### Analyzing the distribution of students by country
According to the chart below, we notice that the country with the most students evaluated was the "UK" and the least was "Canada".

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='country', order=df['country'].value_counts().index)
plt.title('Number of Students by Country')
plt.xlabel('Country')
plt.ylabel('Students')
plt.xticks(rotation=30)

plt.savefig('distribution_of_students_by_country.png', dpi=300)

plt.show()
plt.close()
```

### Proportion of Students Using AI for Academic Purposes by Gender

### AI Adoption by Grade

The code below calculates how many students from each grade level reported using AI tools for academic purposes.

- Only students who use AI (`True`) are considered.
- The results are grouped by grade, showing the total number of AI users per academic year.
- Finally, the grade with the highest AI adoption is identified.

This helps to understand which grade levels have the most engagement with AI tools.


```python
uso_por_ano = df[df['uses_ai_for_study'] == True].groupby('grade')['uses_ai_for_study'].count()

print(uso_por_ano.sort_values(ascending=False))

print(f'\nThe grade level with highest AI tool engagement is: {uso_por_ano.idxmax()}.')
```

### AI Usage by Academic Year

The charts below provide two complementary perspectives on AI adoption across different grade levels:

- **Left Chart (Absolute Count):** Displays the total number of students in each grade who reported using AI tools for study purposes.
- **Right Chart (Proportion):** Shows the proportion of AI users relative to the total number of students in each grade (ranging from 0 to 1).

These visualizations help identify:
- Which academic years have the highest absolute number of AI users.
- In which grades the adoption rate is proportionally higher, regardless of class size.

Combining both perspectives is crucial to avoid misleading conclusions, especially when class sizes vary between grades.


```python
fig = plt.figure(figsize=(14,6))

# Chart 1: Absolute count of AI users by grade level
plt.subplot(1, 2, 1)
sns.countplot(data=df[df['uses_ai_for_study'] == True], x='grade', order=df['grade'].unique())
plt.title('Quantidade de Alunos que Usam IA por Série')

# Chart 2: Proportion of AI usage by grade
plt.subplot(1, 2, 2)
sns.barplot(data=df, x='grade', y=df['uses_ai_for_study'].astype(int), order=df['grade'].unique())
plt.title('Proporção de Alunos que Usam IA por Série')
plt.ylabel('Proporção de Uso (0 a 1)')

plt.tight_layout()
plt.savefig('Figure_Numbers_Proportions_Students_IA_Grade.png', dpi=300)
plt.show()
plt.close()

```

### Correlation Heatmap - Numerical Variables

The heatmap below shows the correlation matrix between the dataset's numerical variables.

**Key Points:**
- Positive values indicate a direct relationship (variables increase together).
- Negative values suggest an inverse relationship (as one increases, the other decreases).
- Values close to 0 imply no significant linear relationship.

This visualization helps identify potential dependencies, hidden patterns, or multicollinearity between AI tool usefulness ratings and other numeric factors.


```python
# Select only numeric columns
corr = df.select_dtypes(include=['float64', 'int64', 'bool']).corr(numeric_only=True)

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={'size':8})
plt.title('Correlation Heatmap (Numerical Variables)')
plt.xticks(rotation=45)

plt.savefig('Numerical_Correlation_Between_Variables.png', dpi=300)
plt.show()
plt.close()
```

### Pairplot - AI Tools Usefulness Ratings

The pairplot below visualizes pairwise relationships between the usefulness ratings of different AI tools.

**Interpretation:**
- Diagonal plots show the distribution (density) for each individual tool's ratings.
- Off-diagonal plots show scatterplots comparing two tools at a time, helping to identify possible correlations or trends.


```python
# Defyning a variable for the chart
pair_usefulness = [col for col in df.columns if col.startswith('usefulness_')]

sns.pairplot(df[pair_usefulness], corner=True, diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot of AI Tools Usefulness Ratings', y=1.02)

plt.savefig('AI_Tools_Usefulness_Ratings.png', dpi=300)
plt.show()
plt.close()

```

### Conclusions

- The majority of students who report being familiar with AI already apply these tools for academic purposes, indicating a direct relationship between familiarity and adoption.
- Students from the United Kingdom exhibit the highest levels of AI awareness and usage frequency among the countries analyzed.
- Non-binary students show slightly different patterns of AI tool usage compared to male and female students, suggesting potential differences in acceptance or access, though further investigation is needed.
- The most widely adopted AI tool among students is Notion AI, followed by ChatGPT, reflecting the growing popularity of generative AI platforms for educational support.
- 9th grade students are the most active in utilizing AI tools, which may correlate with their academic demands or technological curiosity at that educational stage.
- Overall, students perceive AI's impact on education as positive, reinforcing the idea that these technologies are being integrated into learning routines with favorable outcomes.

**Note:** While these trends provide valuable insights, some results may be influenced by sample distribution, missing data, or demographic concentration in specific regions.
